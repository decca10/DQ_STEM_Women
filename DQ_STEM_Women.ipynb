# STEM and Women

This project will explore whether there is a gap between Women in STEM by looking at degrees obtained by women from 1970 to 2012.  I will be using a dataset published by [The Department of Education Statistics](https://nces.ed.gov/programs/digest/2013menu_tables.asp) But, cleaned by [Randal Olson](http://www.randalolson.com/2014/06/14/percentage-of-bachelors-degrees-conferred-to-women-by-major-1970-2012/), a data scientist at the University of Pennsylvania.  

```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline

women_degrees = pd.read_csv('women_degrees.csv')
women_degrees.info()
```

```python
print(women_degrees.iloc[0])
```

## Comparing Charts

I will create charts for a number of STEM related fields and compare those to those considered to be liberal arts.  I will also create a category of "other" to explore any other relationships. And I will add a comparison line at the 50 mark to show which degrees are closer to a 50/50 split.

```python
stem_cats=['Engineering','Computer Science','Psychology','Biology','Physical Sciences','Math and Statistics']
lib_arts_cats = ["Foreign Languages",'English','Communications and Journalism', 'Art and Performance','Social Sciences and History']
other_cats = ["Health Professions",'Public Administration','Education','Agriculture','Business','Architecture']

fig=plt.figure(figsize=(20,20))

#First Column: STEM
for sp in range(0,18,3):
    cat_index=int(sp/3)
    ax=fig.add_subplot(6,3,sp+1)
    cb_dark_blue = (0/255,107/255,164/255) #Women
    cb_orange = (255/255,128/255,14/255) #Men
    cb_gray = (171/255,171/255,171/255)
    ax.plot(women_degrees['Year'],women_degrees[stem_cats[cat_index]],c=cb_dark_blue,label='Women',linewidth=3)
    ax.plot(women_degrees['Year'],100-women_degrees[stem_cats[cat_index]],c=cb_orange,label='Men',linewidth=3)
    for key,spine in ax.spines.items(): #remove the outer grid lines
        spine.set_visible(False)
    ax.set_xlim(1968,2011) #set the x tick spread
    ax.set_ylim(0,100) #set the y tick spread
    ax.set_title(stem_cats[cat_index]) #set title per chart
    ax.tick_params(bottom=False,top=False,right=False,left=False,labelbottom = False) #remove all tick labels
    ax.set_yticks([0,100]) #remove all y tick label except the top and bottom
    plt.subplots_adjust(hspace=0.7) #adjust title spacing
    ax.axhline(50,c=cb_gray,alpha=0.3) #50/50 comparison line at 30% transparency
    if cat_index == 0:
        ax.text(2005,92,"Men")
        ax.text(2005,6,"Women")
    if cat_index == 5:
        ax.text(2005,60,"Men")
        ax.text(2003,33,"Women")
        ax.tick_params(labelbottom=True)

#Second Column: Liberal Arts Degrees
for sp in range(1,16,3):
    cat_index=int((sp-1)/3)
    ax=fig.add_subplot(6,3,sp+1)
    cb_dark_blue = (0/255,107/255,164/255)
    cb_orange = (255/255,128/255,14/255)
    cb_gray = (171/255,171/255,171/255)
    ax.plot(women_degrees['Year'],women_degrees[lib_arts_cats[cat_index]],c=cb_dark_blue,label='Women',linewidth=3)
    ax.plot(women_degrees['Year'],100-women_degrees[lib_arts_cats[cat_index]],c=cb_orange,label='Men',linewidth=3)
    for key,spine in ax.spines.items():
        spine.set_visible(False)
    ax.set_xlim(1968,2011)
    ax.set_ylim(0,100)
    ax.set_title(lib_arts_cats[cat_index])
    ax.tick_params(bottom=False,top=False,right=False,left=False,labelbottom=False)
    ax.set_yticks([0,100])
    plt.subplots_adjust(hspace=0.7)
    ax.axhline(50,c=cb_gray,alpha=0.3)
    if cat_index == 0:
        ax.text(2003,78,"Men")
        ax.text(2003,18,"Women")
    if cat_index==4:
        ax.tick_params(labelbottom=True)

#Third Column: Other Categories
#Third Column: Other Categories
for sp in range(2,20,3):
    cat_index=int((sp-2)/3)
    ax=fig.add_subplot(6,3,sp+1)
    cb_dark_blue = (0/255,107/255,164/255)
    cb_orange = (255/255,128/255,14/255)
    cb_gray = (171/255,171/255,171/255)
    ax.plot(women_degrees['Year'],women_degrees[other_cats[cat_index]],c=cb_dark_blue,label='Women',linewidth=3)
    ax.plot(women_degrees['Year'],100-women_degrees[other_cats[cat_index]],c=cb_orange,label='Men',linewidth=3)
    for key,spine in ax.spines.items():
        spine.set_visible(False)
    ax.set_xlim(1968,2011)
    ax.set_ylim(0,100)
    ax.set_title(other_cats[cat_index])
    ax.tick_params(bottom=False,top=False,right=False,left=False,labelbottom=False)
    ax.set_yticks([0,100])
    plt.subplots_adjust(hspace=0.7)
    ax.axhline(50,c=cb_gray,alpha=0.3)
    if cat_index == 0:
        ax.text(2003,91,"Men")
        ax.text(2003,19,"Women")
    if cat_index == 5:
        ax.text(2006,30, "Men")
        ax.text(2006,62,"Women")
        ax.tick_params(labelbottom=True)

fig.savefig('gender_degrees.png') #save image
plt.show()
```
